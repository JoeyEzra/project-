```python
x = df.drop(columns=['Appliances', 'lights', 'date'])
y = df['Appliances']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Create an instance of the standard scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Trains the model on the training data
model.fit(x_train_scaled, y_train)

# uses the trained model to make Predictions 
model_pred = model.predict(x_train_scaled)
```

```python
from sklearn.metrics import root_mean_squared_error, mean_absolute_error, r2_score
# root mean square error model performance evaluation
model_rmse = root_mean_squared_error(y_train, model_pred)

# mean absolute error model performance evaluation
model_mae = mean_absolute_error(y_train, model_pred) 

```

```python
print(f"The mean absolute error for the training set to three decimal places is: {model_mae:.3f}")
```

```python
print(f"The root mean squared error for the training set to three decimal places is: {model_rmse:.3f}")
```

```python
# Trains the model on the test data
#model.fit(x_test_scaled, y_test)

# uses the trained model of the test data to make Predictions 
test_pred = model.predict(x_test_scaled)
```

```python
# root mean square error model performance evaluation
test_rmse = root_mean_squared_error(y_test, test_pred)

# mean absolute error model performance evaluation
test_mae = mean_absolute_error(y_test, test_pred) 
```

```python
print(f"The mean absolute error for the test set to three decimal places is: {test_mae:.3f}")
```

```python
print(f"The root mean squared error for the test set to three decimal places is: {test_rmse:.3f}")
```

```python
# r2 score to determine overfitting of data
r2_train = r2_score(y_train, model_pred)
r2_test= r2_score(y_test, test_pred)
```

```python
print(f'The r2 score for the training set is: {r2_train:.3f}')
print(f'The r2 score for the test set is: {r2_test:.3f}')

# The model shows a very low performance
```

```python
from sklearn.linear_model import Ridge, Lasso
ridge = Ridge()

# Trains the model on the training data
ridge.fit(x_train, y_train)

# use the trained model to make Predictions
ridge_pred = ridge.predict(x_test)
```

```python
ridge_rmse = root_mean_squared_error(y_test, ridge_pred)
```

```python
print(f'The root mean square for ridge is: {ridge_rmse:.3f}')

# no difference was observed
```

```python
lasso = Lasso()

# Trains the model on the training data
lasso.fit(x_train, y_train)

# use the trained model to make Predictions
lasso_pred = lasso.predict(x_test)
```

```python
lasso_rmse = root_mean_squared_error(y_test, lasso_pred)
```

```python
print(f'The root mean squared for lasso is: {lasso_rmse:.3f}')
```

```python
# Get the feature weights
lasso_weights = lasso.coef_

# Count how many are non-zero
non_zero_count = sum(i != 0 for i in lasso_weights)

print(f"Number of features with non-zero weights: {non_zero_count}")
```
